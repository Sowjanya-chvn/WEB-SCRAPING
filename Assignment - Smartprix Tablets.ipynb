```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

# SMARTPRIX

# DATA COLLECTION

# Pages 1-25

```python
brand=[]
model=[]
processor=[]
ram=[]
rom=[]
battery=[]
ss=[]
prices=[]
for i in range(1,26):
    urls=("https://www.smartprix.com/tablets/most-popular-tablets-pages=")+str(i)
    print(urls)
    page=requests.get(urls)
    soup=BeautifulSoup(page.text)
    
    # 1) Prices
    p=soup.find_all('span',class_='price')
    
    for i in p:
        prices.append(i.text)
    
    # 2) Brand 
    b=soup.find_all('a',class_='name clamp-2')
    
    for i in b:
        brand.append(i.text.split()[0])
    
    # 3) Brand Model
    
    for i in b:
        model.append(i.text.split()[1])
    
    # 4) Processor
    
    c=soup.find_all('ul',class_='sm-feat specs')
    
    for i in c:
        a=i.text
        z1=re.findall(r"(\d+.\d+)\sGHz\sProcessor",a)
        z2=re.findall(r"(\d+)\sGHz\sProcessor",a)
        if len(z1) > 0 and len(z2) > 0:
            processor.append(z1[0])
            processor.append(z2[0])
            
    # 5) ROM   

    for i in c:
        a=i.text
        if "GB" in a:
            z=re.findall(r"(\d+)\sGB\sinbuilt",a)
            if len(z)>0:
                rom.append(z[0])
        elif "TB" in a:
            z=re.findall(r"(\d+)\sTB\sinbuilt",a)
            if len(z)>0:
                rom.append(z[0])
        
    # 6) RAM

    
    for i in c:
        a=i.text
        if "GB" in a:
            z=re.findall(r"(\d+)\sGB\sRAM",a)
            if len(z)>0:
                ram.append(z[0])
        elif "MB" in a:
            y=re.findall(r"(\d+)\sMB\sRAM",a)
            if len(y)>0:
                ram.append(y[0])
        
    
    # 7) Battery
    
    for i in c:
        a=i.text
        z=re.findall(r"(\d+)\smAh",a)
        if len(z)>0:
            battery.append(z[0])
        
    # 8) Screen Size
    
    for i in c:
        a=i.text
        z1=re.findall(r"(\d+.\d+)\sinches",a)
        #z2=re.findall(r"(\d+)\sinches",a)
        if len(z1) > 0:
            ss.append(z1[0])
            #ss.append(z2[0])      and len(z2) > 0

    print("Brands------>",len(brand))
    print("Brand Model------>",len(model))
    print("Prices------>",len(prices))
    print("RAM------>",len(ram))
    print("Battery------>",len(battery))
    print("ss------>",len(ss))
    print("ROM------>",len(rom))
    print("Processor------>",len(processor))

print("__________________")

print("Brands------>",len(brand))
print("Brand Model------>",len(model))
print("Prices------>",len(prices))
print("RAM------>",len(ram))
print("Battery------>",len(battery))
print("ss------>",len(ss))
print("ROM------>",len(rom))
print("Processor------>",len(processor))
```
